```python
#x has all the independent variables  and y has the variable that is to be classified/predicted
x=df.select_dtypes(include=['float64','int64'])
y=df.select_dtypes(include=['object'])
```

```python
#splitting the training and testing data
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
y.value_counts()
# Using SMOTE to see if the values in 'y' is uniformly distributed
import imblearn
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x_smote,y_smote=sm.fit_resample(x,y)
y_smote.value_counts()
```

```python
#Training the set using the smote values
from sklearn.model_selection import train_test_split
x_train_smote,x_test_smote,y_train_smote,y_test_smote=train_test_split(x_smote,y_smote,test_size=0.20)
```

```python
# Normalizing the values of the features in x
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_sc_sm_train=sc.fit_transform(x_train_smote)
x_sc_sm_test=sc.transform(x_test_smote)
```

```python
#using the naive bayes classification algorithm
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_sc_sm_train,y_train_smote)
```

```python
y_predicted=gnb.predict(x_sc_sm_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_smote, y_predicted)
print(cm)
```

```python
#predicting the y
knn_y_pred=KNN.predict(x_sc_sm_test)
knn_y_pred
```

```python
from sklearn.metrics import accuracy_score

print('The accuracy of the KNN with K= is {}%'.format(round(accuracy_score(knn_y_pred,y_test_smote)*100,2)))
```
